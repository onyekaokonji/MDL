```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import random
import cv2
import os

from tensorflow.keras.utils import to_categorical

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score, roc_auc_score, confusion_matrix
from sklearn.metrics import classification_report ,confusion_matrix
from tensorflow.keras.models import load_model, Sequential, Model
from tensorflow.keras.preprocessing import image
import tensorflow
import pickle
from tensorflow.keras.applications import nasnet
from glob import glob
from tensorflow import keras
from tensorflow.keras.layers import Dense, Input, InputLayer, Flatten, Conv2D, Dropout, MaxPooling2D, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint

# import class_activation_map as cam
import preprocessing 
%matplotlib inline
```

```python
base_path = '../input/cell-images-for-detecting-malaria/cell_images'
```

```python
Dataset = preprocessing.load_and_preprocess_images(base_path)
```

```python
Data = pd.DataFrame(Dataset, columns = ['Images', 'Labels'])
```

```python
Data.head()
```

```python
def Labels_to_numeric(x):
        if x == 'Parasitized':
          return 1
        if x == 'Uninfected':
          return 0
Data['Labels'] = Data['Labels'].apply(Labels_to_numeric)
print(Data.head())
```

```python
x = np.array([i for i in Data['Images']]).reshape(-1, 64, 64, 3)
y = to_categorical([i for i in Data['Labels']])
```

```python
len(x)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(x, y, test_size=0.2, shuffle = True, random_state = 2)
```

```python
print(X_train.shape)
print(X_val.shape)
print(Y_train.shape)
print(Y_val.shape)
```

```python
X_train = X_train/255.
X_val = X_val/255.
```

```python
from tensorflow.keras.metrics import Precision, Recall, FalseNegatives, FalsePositives, TruePositives, TrueNegatives
```

```python
# defining evaluation metrics

precision = Precision()
recall = Recall()
false_positive = FalsePositives()
false_negative = FalseNegatives()
true_positive = TruePositives()
true_negative = TrueNegatives()
```

```python
class myCallback(tensorflow.keras.callbacks.Callback):
    def on_epoch_end(self, epochs, logs = {}):
        if (logs.get('accuracy') > 0.98):
            self.model.stop_training = True

myCallback = myCallback()
```

```python
adam = Adam(learning_rate = 0.001)
# sgd = SGD(learning_rate = 0.0001, momentum = 0.8, nesterov = True)
```

### NASNet

```python
model_d = nasnet.NASNetLarge(weights="imagenet", 
                    include_top=False,
                    input_tensor=Input(shape = (64,64,3))
                    )
```

```python
z = model_d.output
z = GlobalAveragePooling2D()(z)
z = BatchNormalization()(z)
z = Dropout(0.5)(z)
z = Dense(1024,activation='relu')(z) 
z = Dense(512,activation='relu')(z) 
z = BatchNormalization()(z)
z = Dropout(0.5)(z)
preds=Dense(2,activation='softmax')(z)
```

```python
nasnet_model = Model(inputs=model_d.input,outputs=preds)
nasnet_model.summary()
```

```python
for layer in model_d.layers:
  layer.trainable = False
```

```python
nasnet_model.compile(optimizer = adam, loss = 'categorical_crossentropy', metrics = ['accuracy', precision, recall, true_positive, true_negative, false_positive, false_negative])
```

```python
nasnet_history = nasnet_model.fit(x = X_train, y = Y_train, validation_data = (X_val, Y_val), epochs = 30, callbacks = myCallback)
```

```python
filepath="./models/weights-improvement-{epoch:02d}-{val_accuracy:.2f}.hdf5"
```

```python
#save the model in .h5 format
nasnet_model.save(filepath = 'nasnet_model', save_format = 'h5', include_optimizer = True)
nasnet_model.save(filepath = 'nasnet_model_tf', save_format = 'tf', include_optimizer = True)
```

```python
# plot the loss
plt.plot(nasnet_history.history['loss'], label='train loss')
plt.plot(nasnet_history.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(nasnet_history.history['accuracy'], label='train acc')
plt.plot(nasnet_history.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')


# plot the Precision
plt.plot(nasnet_history.history['precision'], label='train precision')
plt.plot(nasnet_history.history['val_precision'], label='val precision')
plt.legend()
plt.show()
plt.savefig('Precision')

# plot the Recall
plt.plot(nasnet_history.history['recall'], label='train recall')
plt.plot(nasnet_history.history['val_recall'], label='val recall')
plt.legend()
plt.show()
plt.savefig('Recall')
```

```python
hist_df = pd.DataFrame(nasnet_history.history)
hist_json_file = './nasnet_history.json'
with open(hist_json_file, mode = 'w') as f:
  hist_df.to_json(f)
```
